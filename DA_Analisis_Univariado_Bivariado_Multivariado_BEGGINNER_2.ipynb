Para este primer caso usaremos el dataset llamado "tips" includio en la biblioteca Seaborn, este conjunto de datos contiene información sobre propinas en diferentes comidas incluyendo el total de la factura, sexo, cliente fumador o no, dia, tiempo y tamaño de la fiesta.

```python
#importar librerias
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
#Dar estilo a seaborn
sns.set(style="darkgrid")

#cargar el data set
df = sns.load_dataset('tips')
df.head()
```

- Siempre que nos llega un dataset tenemos que identificar que variable queremos predecir o como podemos encontrar relaciones entre unas y otras, es util obtener información basica sobre el conjunto de datos para lo que utilisaremos: ".info()" y ".describe()"

```python
df.info()
```

El ".info()" nos muestra en este caso que tenemos 244 casos, así como el nombre de las columnas, si hay o no nulos y el tipo de dato que hay en cada columna

```python
df.describe()
```

- En ".describe()" nos muestra datos estadisticos como la media  de una cuenta completa, la propina promedio, así como el maximo y minimo, desviacion estandar entre otras como los percentiles

## ANALISIS UNIVARIADO

Este analisis implica analizar una variable a la vez, podemos usar histogramas, diagramas de caja y graficos de densidad para observar la distribución de una variable, esto nos da información acerca de como es la variacion y comportamiento de esta variable.

GRÁFICO #1

```python
#Construcion del histograma de "total_bill"
sns.histplot(data=df, x='total_bill')
plt.plot()
plt.show()
```

- Este histograma nos genero una distribución de la cuenta total de las cuentas que hay, asi que, observamos aqui que la mayor frecuencia en las cuentas a pagar recae entre los 10 y 25 dolares

GRÁFICO #2

```python
sns.boxenplot(data=df, x='total_bill')
plt.plot()
```

- Aqui observamos que en esta gráfica de boxplot se ve de una manera mas clara la frecuencia entre el rango de 10 y 25 donde encontramos la mayoria de datos, asi como podemos observar los valores atipicos estos estan fuera de nuestro analisis

GRÁFICO #3

```python
sns.kdeplot(data=df, x='total_bill')
plt.show()
```

- En este tercer gráfico podemos ver la densidad y total_bill, asi como en los dos gráficos anteriores se muestra el rango donde mas hay frecuencia

- Estos 3 gráficos son muy utiles en el Analisis Univariado

- CONCLUSIÓN: En base al analisis hecho con el dataset llamado "Tips" perteneciente a Seaborn podemos destacar lo siguiente:

  Analisamos la variables "total_bill" la cual nos dice los siguiente:

  En el "gráfico 1" nos muestra que el rango de dolares que se pagan en cuentas recae entre los 10 y 20 dolares, asi mismo complementamos entre analisis con el "gráfico 2"  y "gráfico 3" donde visualizamos de una manera más clara este frecuencia de los datos y observamos tambien los valores atipicos que se presentan en el analisis.


##ANALISIS BIVARIADO

El analisis bivariado se centra en estudiar las relaciones entre dos variables podemos usar gráficos de dispersion, gráficos de barras.

- Comenzaremos analizando la relación entre la propina y el total de la cuenta pagado

GRÁFICO #1

```python
#Gráfico de disperscion
sns.scatterplot(data = df, x ='total_bill', y='tip')
plt.show()
```

- Como podemos observa es un gráfico con mucha disperción pero si hay una tendencia en que mientras mas grande es la cuenta la propina tambien y la mayor concentracion se da en un rango de  de cuentas de 10 a 25 dolares las propinas estan en un rango de 2 a 5 dolares

- Observando el gráfico podemos determinar que tipo de modelos pueden ser apliacados en estos casos

GRÁFICO #2

```python
#Gráfico de barras
sns.barplot(data = df, x ='sex', y ='tip')
plt.show()
```

- Acorde a este gráfico de barras podemos observar que el tipo de clientes da la menor y mayor propina, es decir, en este caso el sexo masculino tiende a dar una mayor propina comparado con el sexo femenino

GRÁFICO #3

```python
sns.boxenplot(data = df, x = 'sex', y ='tip')
plt.show()
```

*   Con el gráfico 3 podemos observar más claramente la variable "tips" que hace referencia a la propina dada por ambos sexos en este gráfico notamos que la frecuencia de propina del sexo masculino ronda entre los 2 a 4 dolares mientras que para el caso del sexo femenico ronda entre los 2 y 3 dolares, asi mostrando datos atipicos donde para el sexo masculino sobre pasa propinas e 6 a 10 dolares como máximo mientras el sexo femenino sobrepasa propinas de 5 a 6 dolares siendo esta la más alta.

GRÁFICO #4

```python
#sexo vs total de la cuenta
sns.boxenplot(data = df, x = 'sex', y ='total_bill')
plt.show()
```

- Podemos observar en el gráfico que el sexo masculino tiende a pagar mas, estando en un rango de 15 a 25 dolares aproximadamente a comparación del sexo femenino que esta en un rango de 15 a 20 dolares aproximadamente

- CONCLUSIÓN: En base al analisis hecho con el dataset llamado "Tips" perteneciente a Seaborn podemos destacar lo siguiente:

  1) Analisamos la variables "total_bill" con respecto a la variable "tip" en donde con el gráfico 1 se pudo observar una tendencia en que mientras mas grande es la cuenta la propina tambien y la mayor concentracion se da en un rango de cuentas de 10 a 25 dolares mientras que las propinas estan en un rango de 2 a 5 dolares.

  2) Con el "gráfico 2" analizamos las variables "sex" y "tip" y con ayuda del "gráfico 3" complementamos observando  más claramente la variable "tips" que hace referencia a la propina dada por ambos sexos en este gráfico notamos que la frecuencia de propina del sexo masculino ronda entre los 2 a 4 dolares mientras que para el caso del sexo femenico ronda entre los 2 y 3 dolares, asi mostrando datos atipicos donde para el sexo masculino sobre pasa propinas e 6 a 10 dolares como máximo mientras el sexo femenino sobrepasa propinas de 5 a 6 dolares siendo esta la más alta.

## ANALISIS MULTIVARIADO

El analisis multivariado implica estudiar más de dos variables al mismo tiempo, podemos usar gráficos de dispersion en pares, gráficos de cajas, gráficos de correlación

GRÁFICO #1

```python
sns.pairplot(df, hue="sex")
plt.show()
```

*   Como podemos obervar aqui se hizo una compración de la variable "sex" con respecto a todas las demas variables de nuestro dataset mientras sean numericas

*   CONCLUSIÓN: En el "gráfico 1" pairplot se comparo la variable "sex" donde se puede concluir lo siguiente:

1) Con respecto a la variable "total_bill" y la variable "tip" se oberva que el sexo masculino tiende a dar una mayor porpina con una frecuencia de 2 a 5 dolares

2) Con respecto a la variable "total_bil" y la variable "size" se oberva una mayor frecuencia en el sexo masculino el cual suele pagar cuentas de 20 a 40 dolares con un rango de 3 a 4 personas

3) Con respecto a la variable "tip" y la variable "size" se observa que hay una mayor frecuencia de propinas entre los 3 a 6 dolares dadas por mesas de 3 a 4 personas y esta tendencia se oberva mayormente en el sexo masculino

4) Con respecto a la variable "tip" y la variable "tip" se observa que hay una mayor frecuencia de propinas dadas por el sexo masculino

5) Con respecto a la variable "size" y la variable "size" se oberva que suele haber una mayor frecuencia del sexo masculino con mesas de 2 personas que acomparacion del sexo femenino tiene una mayor frecuencia de 2 personas, es decir, el sexo masculno suele salir mas frecuentemente en pares que el femenino
